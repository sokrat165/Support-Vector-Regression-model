## Normalization 

###### Normalization is an important preprocessing step in machine learning to scale features to a similar range. You can use ''  scikit-learn's StandardScaler  '' to perform normalization on your data

after normalizing the data using the StandardScaler, it is converted to a NumPy array which does not have a 'columns' attribute like a pandas DataFrame.

If you want to convert the normalized data back to a DataFrame while preserving column names, you can do the following:

```python
# Initialize the StandardScaler
scaler = StandardScaler()
# Normalize the DataFrame columns
normalized_data = scaler.fit_transform(df)
# Get the column names after normalization
print(type(normalized_data))

normalized_columns = scaler.get_feature_names_out(input_features=df.columns)

#conver data type from array to dataframe to use function colums to named colum from datafram
# Create a DataFrame with the normalized data and column names

normalized_df = pd.DataFrame(normalized_data,columns=normalized_columns)
print(normalized_df)
```

```python
x=df.values[:,:2]
y=df.values[:,2]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=20)
print(y_train)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


# Create an SVR model
regressor = SVR(kernel='linear', C=2)

# Train the model
regressor.fit(x_train, y_train)

# Make predictions on the test set
y_pred = regressor.predict(x_test)

# mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python


# Assuming x_train, x_test, y_train, y_test are defined

# Create and fit the Support Vector Regression model
regressor = SVR(kernel='linear')
regressor.fit(x_train, y_train)

# Create a meshgrid for visualization
xx, yy = np.meshgrid(np.linspace(min(x_train[:, 0]), max(x_train[:, 0]), 100),
                     np.linspace(min(x_train[:, 1]), max(x_train[:, 1]), 100))
xy_pairs = np.column_stack((xx.ravel(), yy.ravel()))
zz = regressor.predict(xy_pairs)
zz = zz.reshape(xx.shape)

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_test[:, 0], x_test[:, 1], y_test, c='blue', marker='o', label='Test Data')
ax.scatter(x_train[:, 0], x_train[:, 1], y_train, c='red', marker='o', label='Test Data')

ax.plot_surface(xx, yy, zz, cmap='viridis', alpha=0.5, label='Regression Plane')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')
ax.set_title('Support Vector Regression in 3D')
ax.legend()
plt.show()
```
